```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
bd = "bank-full.csv"
names = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact','day','month','duration',
        'campaign','pdays','previous','poutcome','y']
data = pd.read_csv(bd, names=names)
```

```python
soltero = pd.DataFrame()
soltero = soltero.append(data[(data['age']<20) & (data['marital']=='single') ][['campaign']].mean(),ignore_index=True)
soltero = soltero.append(data[((data['age']>=20) & (data['age']<30)) & (data['marital']=='single')][['campaign']].mean(),ignore_index=True)
soltero = soltero.append(data[((data['age']>=30) & (data['age']<40)) & (data['marital']=='single')][['campaign']].mean(),ignore_index=True)
soltero = soltero.append(data[((data['age']>=40) & (data['age']<50)) & (data['marital']=='single')][['campaign']].mean(),ignore_index=True) 
soltero = soltero.append(data[((data['age']>=50) & (data['age']<60)) & (data['marital']=='single')][['campaign']].mean(),ignore_index=True)
soltero = soltero.append(data[((data['age']>=60) & (data['age']<70)) & (data['marital']=='single')][['campaign']].mean(),ignore_index=True)
soltero = soltero.append(data[(data['age']>=70) & (data['marital']=='single')][['campaign']].mean(),ignore_index=True)

casado= pd.DataFrame()
casado = casado.append(data[(data['age']<20) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True)
casado = casado.append(data[((data['age']>=20) & (data['age']<30)) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True)
casado = casado.append(data[((data['age']>=30) & (data['age']<40)) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True)
casado = casado.append(data[((data['age']>=40) & (data['age']<50)) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True) 
casado = casado.append(data[((data['age']>=50) & (data['age']<60)) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True)
casado = casado.append(data[((data['age']>=60) & (data['age']<70)) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True)
casado = casado.append(data[(data['age']>=70) & (data['marital']=='married')][['campaign']].mean(),ignore_index=True)

divorciado= pd.DataFrame()
divorciado = divorciado.append(data[(data['age']<20) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True)
divorciado = divorciado.append(data[((data['age']>=20) & (data['age']<30)) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True)
divorciado = divorciado.append(data[((data['age']>=30) & (data['age']<40)) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True)
divorciado = divorciado.append(data[((data['age']>=40) & (data['age']<50)) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True) 
divorciado = divorciado.append(data[((data['age']>=50) & (data['age']<60)) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True)
divorciado = divorciado.append(data[((data['age']>=60) & (data['age']<70)) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True)
divorciado = divorciado.append(data[(data['age']>=70) & (data['marital']=='divorced')][['campaign']].mean(),ignore_index=True)
```

```python
rango_edades = '<30', '35', '45', '55', '60','65','70>'
plt.plot()
plt.plot(rango_edades,soltero,'b--o',label='single')
plt.plot(rango_edades,casado,'y--o',label='single')
plt.plot(rango_edades,divorciado,'g--o',label='single')
plt.title('contactos promedio por rango de años ')
plt.xlabel('rango de años')
plt.ylabel('promedio de contactos')
plt.legend()
plt.show()
```
