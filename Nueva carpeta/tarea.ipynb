```python
import pandas as pd
import numpy as np

bd = "bank-additional-full.csv"
names = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan', 'contact','month','day_of_week',
        'duration','campaign','pdays','previous','poutcome','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m',
        'nr.employed','y']
data = pd.read_csv(bd, names=names)
```

```python
# age: edad de entrevistado
# job: tipo de ocupacion "admin.","blue-collar","entrepreneur","housemaid","management","retired","self-employed","services","student","technician","unemployed","unknown"
# marital: estado civel "divorced","married","single","unknown"
#education: grado de educacion "basic.4y","basic.6y","basic.9y","high.school","illiterate","professional.course","university.degree","unknown"
#default:  tiene el crédito por defecto? "yes","no"
#housing: tiene prestamo de vivienda "yes","no"
#loan: tuene prestamo personal? "yes","no"
#contact: comunicación de contacto "telephone","cellular"
#month: último contacto mes del año "jan", "feb", "mar", ..., "nov", "dec"
#day_of_week: last contact day of the week (categorical: "mon","tue","wed","thu","fri")
#duration: Duración del último contacto, en segundos. numero
#campaign: cantidad de veces contactado, numero
#pdays: la cantidad de días que pasaron después de que el cliente 
#fue contactado por última vez desde una campaña anterior 
#(numérico, -1 significa que el cliente no fue contactado previamente)
#previous: número de contactos realizados antes de esta campaña y para este cliente (numérico)
#poutcome: Resultado de la campaña de marketing anterior. "unknown","other","failure","success"
#emp.var.rate: tasa de variación del empleo - indicador trimestral (numérico)
#cons.price.idx: índice de precios al consumidor - indicador mensual (numérico)
#cons.conf.idx: consumer confidence index - monthly indicator (numeric) 
#euribor3m: Euribor tasa de 3 meses - indicador diario (numérico)
#nr.employed: número de empleados - indicador trimestral (numérico)
#y: ¿El cliente ha suscrito un depósito a plazo? "yes","no"
data
```
