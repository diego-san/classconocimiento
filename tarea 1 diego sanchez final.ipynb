```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.io as sio


bd = "bank-full.csv"
names = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact','day','month','duration',
        'campaign','pdays','previous','poutcome','y']
data = pd.read_csv(bd, names=names)
```

```python
# age: edad de entrevistado
# job: tipo de trabajo "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
# marital: estado civel "married","divorced","single"
#education: grado de educacion "unknown","secondary","primary","tertiary"
#default:  tiene el crédito por defecto? "yes","no"
#balance: saldo medio anual numero tipo de dato
#housing: tiene prestamo de vivienda "yes","no"
#loan: tiene prestamo personal? "yes","no"
#contact: comunicación de contacto "unknown","telephone","cellular"
#day: último día de contacto del mesn numeros de dias
#month: último contacto mes del año "jan", "feb", "mar", ..., "nov", "dec"
#duration: Duración del último contacto, en segundos. numero
#campaign: cantidad de veces contactado, numero
#pdays: la cantidad de días que pasaron después de que el cliente 
#fue contactado por última vez desde una campaña anterior 
#(numérico, -1 significa que el cliente no fue contactado previamente)
#previous: número de contactos realizados antes de esta campaña y para este cliente (numérico)
#poutcome: Resultado de la campaña de marketing anterior. "unknown","other","failure","success"
#y: has the client subscribed a term deposit? (binary: "yes","no")
```
